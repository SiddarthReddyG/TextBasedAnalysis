### USING REGEX

 Regular Expression, is a sequence of characters that forms a search pattern. RegEx can be used to check if a string contains the specified search pattern.

```python
#Using REGEX


import regex as re

def tokenize(text):
    return re.findall(r'[\w-]*[a-z|A-Z][\w-]*',text)
print(tokenize)
```

### TOKENIZATION

```python
#Tokenization process


testtext = "Let's defeat SARS-CoV-2 together in 2020! www.cdc.gov"
testtokens = tokenize(testtext)
 
print(testtokens)
```

### USING NLTK PACKAGE

The Natural Language Toolkit (NLTK) is a platform used for building Python programs 
that work with human language data for applying in statistical natural language processing (NLP).
It contains text processing libraries for tokenization, parsing, classification, stemming, tagging and semantic reasoning.

```python
#Importing the nltk package


import nltk 
nltk.download('stopwords')
stopwords = set(nltk.corpus.stopwords.words('english'))
print(stopwords)
```

## REMOVING STOPWORDS

```python
#Removing stop words and testing it

def remove_stop(tokens):
    return [t for t in tokens if t.lower() not in stopwords]
cleantokens= remove_stop(testtokens)
print(cleantokens)
```

### PIPELINE AND PREPARE FUNCTION

```python
#Creating a pipeline and prepare function

pipeline = [str.lower, tokenize, remove_stop]

def prepare(text, pipeline):
    tokens = text
    for fun in pipeline:
        tokens = fun(tokens)
    return tokens
```

### PIPELINES

```python
#Using the pipeline

df['tokens'] = df['text'].apply(prepare, pipeline=pipeline)
```

### TOKEN COUNT AND TAIL METHOD

```python
#token counting and tail outputs last rows

df['num_tokens'] = df['tokens'].map(len)
df.tail()# returns last rows
```

### USING COUNTER 

A Counter is a dict subclass for counting hashable objects.

```python
#Frequency

from collections import Counter

tokens = tokenize("She likes my cats and my cats like my sofa.")

testcount= Counter(tokens)
print(testcount)
```

### EXPANDING TOKEN KNOWLEDGE

```python
#Additonal new data


more_tokens = tokenize("She likes dogs and cats.")
testcount.update(more_tokens)
print(testcount)

```

### CHANGE IN FREQUENCY

```python
#Frequency after changes

counter = Counter()
df['tokens'].map(counter.update)
print(counter.most_common(5))
```

### SHIFTING THROUGH DATA

```python
#Adding to df and Less Frequent words are eliminated

freq_df = pd.DataFrame.from_dict(counter, orient='index', columns=['freq'])
freq_df = freq_df.query('freq >= 1')
freq_df.index.name = 'token'
```

### FREQUENCY 

```python
#Show frequency


freq_df = freq_df.sort_values('freq', ascending=False)
freq_df.head(10)
```

### DESCRIPTIVE STATISTICS

```python
#Statitics again

freq_df.describe()
```

### WORDCLOUD EXAMPLES

Word cloud is one of the most powerful and straightforward visualization methods when it comes to text data. The size of words are dependent on the occurrence frequency

```python
#Wordcloud i.e image like summary

from wordcloud import WordCloud
from matplotlib import pyplot as plt

wc = WordCloud(background_color='black', width = 600, height=600, max_words=100, margin=2)

text = df.query("year==2015 and country=='USA'")['text'].values[0]

plt.figure(figsize=(14, 10)) 

wc.generate(text)# Generating the WordCloud based on the extracted text
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
```
